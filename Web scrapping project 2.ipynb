```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://en.wikipedia.org/wiki/List_of_cities_by_number_of_billionaires"
page = requests.get(URL)
soup = BeautifulSoup(page.text,"html")
```

```python
print(soup.prettify)
```

```python
tag_object = soup.title
print(tag_object)
```

```python
print(type(tag_object))
```

```python
tag_object = soup.h1
tag_object
```

```python
if tag_object:
    tag_child = tag_object.b
    if tag_child:
        print(tag_child.text)
    else:
        print("No tag be found within <h1>")

else:
    print ("No <h1> tag found")
    
```
