`````python
#Adding columns with filters using Pandas command apply(lambda)
bios1["Heightcateg"]=bios1["height_cm"].apply(lambda x: n/a if x== 0 else("short" if x<165 else ("Average" if x < 180 else "Tall")))
bios1[["name","Heightcateg"]]
`````

`````python
#we can analyze multiple data using functions
def buildathlete(row):
    if row["Heightcateg"] == "Tall" and row["weight_kg"]>95:
        return "Heavybuild"
    
    elif row["Heightcateg"] == "Average" and row["weight_kg"]>90:
        return "Heavybuild"
    
    elif row["Heightcateg"] == "short" and row["weight_kg"]>85:
        return "Heavybuild"
    
    else:
        return "Average"
    
bios1["Buildathlete"]= bios1.apply(buildathlete, axis=1)
`````
